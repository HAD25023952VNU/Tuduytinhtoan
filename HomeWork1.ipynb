 Bài 2: Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2 chữ số phần thập phân. 

```python
a, b = map(float, input().split())
S = a*b - (3.14*a**2)/4
print(f"{S:.2f}")
```

Bài 3: Viết chương trình nhập vào một “kí tự” c, nếu nó là chữ hoa thì in ra chữ thường tương ứng và ngược lại nếu nó là chữ thường thì in ra chữ hoa tương ứng của nó. (xem trước cấu trúc lựa chọn: if….else, ma) 

```python
c=input()
if "a" <= c <= "z":
    print(c.upper())
elif "A" <= c <= "Z":
    print(c.lower())    
else:
    print("Khong phai chu cai")    
```

Bài 4: Viết chương trình nhập vào một “kí tự”, Kiểm tra xem c có phải là kí tự chữ cái 
không (hoa hoặc thường). Nếu đúng thì in ra “{c} là kí tu alphabet”, nếu sai in ra “{c} 
không phải là kí tự alphabet. 

```python
c = input("nhap vao 1 ki tu")
if "a" <= c <= "z" or "A" <= c <= "Z":
    print(f"{c} là kí tu alphabet")
else: 
    print(f"{c} không phải là kí tự alphabet")
```

Bài 5: Nhập vào chữ cái hoa A, in ra chữ cái thường liền trước chữ cái a thường tương ứng của A. Xét TH khi c = ‘A’ là đặc biệt. 

```python
n = input().upper()
if n == "A":
    print("khong ton tai")
else:
    m = chr(ord(n) - 1)
    print(m.lower())

```

Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không, 
nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó 
không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”. 

```python
import math 
a, b, c = map(float, input().split())
if a+b > c and a+c > b and b+c > a:
    C = (a + b +c)/2
    S = math.sqrt(C*(C-a)*(C-b)*(C-c))
    print(f"{S:.1f}")
else:
    print("Khong phai 3 canh cua tam giac")    
```

```python
a, b, c = map(float, input().split())
if a+b > c and a+c > b and b+c > a:
    C = (a + b +c)/2
    S = (C*(C-a)*(C-b)*(C-c))**0.5
    print(f"{S:.1f}")
else:
    print("Khong phai 3 canh cua tam giac")    
```
